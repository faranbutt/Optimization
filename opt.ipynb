```python
x1 = [x for x in range(100)]
x2 = [x for x in range(1,200,2)]
x1 = np.array(x1)
x2 = np.array(x2)
```

```python
y = 10*x1+100*x2
```

```python
plt.figure(figsize = (10,4))
plt.plot(x1,y)
plt.plot(x2,y)
```
